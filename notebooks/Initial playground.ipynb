```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Reading the data

```python
df = pd.read_json("../data/data.json")
```

```python
# convert timestamp to date and get hour
df['date'] = pd.to_datetime(df['measuredAt'])
df['hour'] = df.date.dt.hour
```

## Plotting temperature against humidity by hour

```python
fig, ax = plt.subplots()
df.plot.hexbin(x='humidity', y='temperature', C='hour',colormap='viridis',gridsize=25,figsize=(16, 12), ax = ax)
plt.xlabel('humidity')
plt.ylabel('temperature')
plt.show()
```

```python
fig, ax = plt.subplots()
df.plot.scatter(x='humidity', y='temperature', c='hour',colormap='viridis',figsize=(16, 12), ax=ax)   
plt.xlabel('humidity')
plt.ylabel('temperature')
plt.show()
```
